```python
import json
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import numpy as np
```

```python
train_file=f'./malaria/training.csv'
test_file = './malaria/test.csv'
image_path = './malaria/images/'
```

Counting The dataset unique values.

```python
train_data = pd.read_csv(train_file)
test_data = pd.read_csv(test_file)

train_len = len(train_data['path'].unique())
print('Total number of Images in train dataset:', train_len)

test_len = len(test_data['path'].unique())
print('Total number of Images in test dataset:', test_len)
```

```python
# Get the list of image files in the directory
image_files = os.listdir(image_path)

# Load the first image in the list
first_image_path = os.path.join(image_path, image_files[0])
img = Image.open(first_image_path)

# Summarize some details about the image
print('Image format:', img.format)
print('Image mode:', img.mode)
print('Image size:', img.size)

# Load image as a pixel array using matplotlib
img_array = mpimg.imread(first_image_path)

# Summarize shape of the pixel array
print('Image datatype:', img_array.dtype)
print('Image shape:', img_array.shape)

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
def subplot(nrows, ncols, data_path):
    # Get the list of image files in the directory
    image_files = os.listdir(data_path)

    # Create a subplot
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10, 10))

    # Loop through the images and plot them
    for i in range(nrows):
        for j in range(ncols):
            # Load the image
            img_path = os.path.join(data_path, image_files[i * ncols + j])
            img = Image.open(img_path)

            # Plot the image
            axes[i, j].imshow(img)
            axes[i, j].axis('off')

    plt.show()

# Example usage
data_path = image_path
subplot(3, 3, data_path)
```

```python
train_categories = train_data['category'].value_counts()
test_categories = test_data['category'].value_counts()

print("Train Categories:")
print(train_categories)

print("\nTest Categories:")
print(test_categories)
```

```python
# Plot for Train Categories
plt.figure(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(train_categories)))
plt.bar(train_categories.index, train_categories.values, color=colors, width=0.6)
plt.title('Distribution of Categories in Train Dataset')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Plot for Test Categories
plt.figure(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(test_categories)))
plt.bar(test_categories.index, test_categories.values, color=colors, width=0.6)
plt.title('Distribution of Categories in Test Dataset')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
